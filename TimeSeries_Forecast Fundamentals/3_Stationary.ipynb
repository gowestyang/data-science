## Variance Stabilization: Box-Cox Transformation

```python
import numpy as np
import pandas as pd

dfItemA = pd.read_csv('data/time_series_daily_A.csv')
seSalesA = dfItemA['sales_qty']

seSalesA.plot()
```

```python
from scipy.stats import boxcox

seSalesA_pos = seSalesA.copy()
seSalesA_pos[seSalesA_pos==0] = 0.00001

arSalesA_bc, op_lambda = boxcox(seSalesA_pos)
print(f"Optimized lambda = {op_lambda}")
pd.Series(arSalesA_bc).plot()
```
